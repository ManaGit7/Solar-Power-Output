```python
# Importing necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Displaying the dataset
df = pd.read_csv('dataset/solarpowergeneration.csv')
df.head()
```

```python
# Check for duplicate values
df.duplicated().sum()
```

```python
? Plot distribution of Power
plt.figure(figsize=(10,6))
sns.histplot(df['generated_power_kw'],bins=30, kde=True)
plt.title('Distribution of Generated Power (kW)')
plt.xlabel('Generated Power (kW)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Bivariate analysis 
# Scattter plot with target feature
plt.figure(figsize=(15, 30))
for i, column in enumerate(df.columns):
    plt.subplot(7, 3, i+1)
    plt.scatter(df[column], df['generated_power_kw'])
    plt.title(f'{column} vs GeneratedPowe(kW)')
    plt.xlabel(column)
    plt.ylabel('GeneratedPowe (kW)')
plt.tight_layout()
plt.show()
```

```python
df.corr()
```

```python
plt.figure(figsize=(15, 10))
sns.heatmap(df.corr(), cmap='coolwarm', annot=True, fmt='.2f')
plt.show()
```

```python
# Outlier
plt.figure(figsize=(15, 30))
for i, column in enumerate(df.columns):
    plt.subplot(7, 3, i+1)
    sns.boxplot(df[column])
plt.show()
```
